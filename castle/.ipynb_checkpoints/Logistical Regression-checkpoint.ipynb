```python
# plot the linear plot of differerent C from 0.001 to 0.1 and penalty
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler

accuracy_list = [[],[]]
accuracy_best = 0
parameter_best = {}
penalty_list = ['l1', 'l2']
labels = ['l1_test', 'l2_test', 'l1_train', 'l2_train']
color = ['red', 'green', 'blue', 'yellow']
c_list = np.linspace(0.001, 0.01, 300)
# tune the hyperparameters
for c in c_list:
    models_list = []
    for i in range(2):
        models_list.append(LogisticRegression(penalty=penalty_list[i], C=c, solver='liblinear'))
        
    for i in range(2):
        models_list[i].fit(x_train,y_train)
        accuraccuracy = accuracy_score(y_test, models_list[i].predict(x_test))
        accuracy_list[i].append(accuraccuracy)
        if accuraccuracy > accuracy_best:
            accuracy_best = accuraccuracy
            parameter_best['penalty'] = penalty_list[i]
            parameter_best['C'] = c

        
for i in range(2):
    plt.plot(c_list, accuracy_list[i], label = labels[i], color = color[i])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.legend()
print(accuracy_best)
print(parameter_best)
```
